```python
import numpy as np
import re
import struct

def custom_bytes_to_floats(b):
    # 1. Разбиваем по паттерну ";\\x03" — это явно разделитель между числами
    # удаляем всё лишнее кроме значений float
    parts = re.split(b";\x03", b)

    floats = []
    for p in parts:
        # ищем все 4-байтные куски, кратные float32
        for i in range(0, len(p) - 4, 4):
            chunk = p[i:i+4]
            if len(chunk) == 4:
                try:
                    val = struct.unpack("<f", chunk)[0]  # little endian float32
                    if np.isfinite(val):
                        floats.append(val)
                except Exception:
                    pass
    return np.array(floats, dtype=np.float32)

# Тест на одном примере
b = sample_df['embedding'].iloc[0]
emb = custom_bytes_to_floats(b)
print("Эмбеддинг восстановлен, форма:", emb.shape)
print("Первые 10 значений:", emb[:10])
```

```python
import struct
import numpy as np

def decode_fixed(b):
    floats = []
    # Начинаем с байта 1, идём блоками по 8 байт (первые 4 — мусор, следующие 4 — float)
    for i in range(1, len(b) - 4, 8):
        try:
            val = struct.unpack('<f', b[i:i+4])[0]  # little endian float32
            if np.isfinite(val):
                floats.append(val)
        except:
            pass
    return np.array(floats, dtype=np.float32)

# Тест
b = sample_df['embedding'].iloc[0]
emb = decode_fixed(b)
print("Эмбеддинг восстановлен, форма:", emb.shape)
print("Первые 10 значений:", emb[:10])
print("Мин/макс:", emb.min(), emb.max())
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Путь к модели
model_name = "cointegrated/rubert-tiny2"

# Загружаем токенизатор и модель
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Переводим модель в eval режим (не обучаем)
model.eval()
```

```python
def get_embedding(text):
    """
    Получает эмбеддинг текста через rubert-tiny2
    """
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        # Берем CLS-токен как эмбеддинг
        emb = outputs.last_hidden_state[:,0,:].squeeze(0)
    return emb.numpy()
```

```python
# Декодируем весь столбец
sample_df['model_text_decoded'] = sample_df['model_text'].apply(lambda x: x.decode('utf-8'))

# Проверим
sample_df['model_text_decoded'].head()

sample_texts = sample_df['model_text_decoded'].iloc[:3].tolist()

for i, t in enumerate(sample_texts):
    emb = get_embedding(t)
    print(f"Текст {i}:", t[:50], "...")
    print(f"Эмбеддинг {i} форма:", emb.shape)
    print("Первые 5 элементов:", emb[:5], "\n")
```

```python
import numpy as np
import torch
from tqdm import tqdm

# Количество текстов для генерации эмбеддингов (можно уменьшить/увеличить)
num_samples = 1000

# Берем случайную подвыборку
texts = sample_df['model_text_decoded'].sample(num_samples, random_state=42).tolist()

# Функция получения эмбеддингов
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        emb = outputs.last_hidden_state[:,0,:].squeeze(0)  # CLS токен
    return emb.numpy()

# Генерация эмбеддингов с прогрессбаром
embeddings = []
for t in tqdm(texts, desc="Генерация эмбеддингов"):
    emb = get_embedding(t)
    embeddings.append(emb)

embeddings = np.array(embeddings, dtype=np.float32)
print("Форма итоговой матрицы эмбеддингов:", embeddings.shape)

# Сохраняем для дальнейшей кластеризации
np.save("./data/embeddings_sample.npy", embeddings)
print("Эмбеддинги сохранены в ./data/embeddings_sample.npy")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Пример: сравниваем первые 5 эмбеддингов
sim_matrix = cosine_similarity(embeddings[:5])
print(sim_matrix)
```

```python
from sklearn.cluster import KMeans

n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)
print("Метки KMeans:", labels[:10])
```

```python
import umap
import plotly.express as px
import pandas as pd

reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
emb_2d = reducer.fit_transform(embeddings)

viz_df = pd.DataFrame({
    "x": emb_2d[:,0],
    "y": emb_2d[:,1],
    "kmeans": labels,
    "text": texts
})

fig = px.scatter(viz_df, x='x', y='y', color='kmeans', hover_data=['text'])
fig.show()
```
