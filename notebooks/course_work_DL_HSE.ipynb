```python
import numpy as np
import re
import struct

def custom_bytes_to_floats(b):
    # 1. –†–∞–∑–±–∏–≤–∞–µ–º –ø–æ –ø–∞—Ç—Ç–µ—Ä–Ω—É ";\\x03" ‚Äî —ç—Ç–æ —è–≤–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –º–µ–∂–¥—É —á–∏—Å–ª–∞–º–∏
    # —É–¥–∞–ª—è–µ–º –≤—Å—ë –ª–∏—à–Ω–µ–µ –∫—Ä–æ–º–µ –∑–Ω–∞—á–µ–Ω–∏–π float
    parts = re.split(b";\x03", b)

    floats = []
    for p in parts:
        # –∏—â–µ–º –≤—Å–µ 4-–±–∞–π—Ç–Ω—ã–µ –∫—É—Å–∫–∏, –∫—Ä–∞—Ç–Ω—ã–µ float32
        for i in range(0, len(p) - 4, 4):
            chunk = p[i:i+4]
            if len(chunk) == 4:
                try:
                    val = struct.unpack("<f", chunk)[0]  # little endian float32
                    if np.isfinite(val):
                        floats.append(val)
                except Exception:
                    pass
    return np.array(floats, dtype=np.float32)

# –¢–µ—Å—Ç –Ω–∞ –æ–¥–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ
b = sample_df['embedding'].iloc[0]
emb = custom_bytes_to_floats(b)
print("–≠–º–±–µ–¥–¥–∏–Ω–≥ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω, —Ñ–æ—Ä–º–∞:", emb.shape)
print("–ü–µ—Ä–≤—ã–µ 10 –∑–Ω–∞—á–µ–Ω–∏–π:", emb[:10])
```

```python
import struct
import numpy as np

def decode_fixed(b):
    floats = []
    # –ù–∞—á–∏–Ω–∞–µ–º —Å –±–∞–π—Ç–∞ 1, –∏–¥—ë–º –±–ª–æ–∫–∞–º–∏ –ø–æ 8 –±–∞–π—Ç (–ø–µ—Ä–≤—ã–µ 4 ‚Äî –º—É—Å–æ—Ä, —Å–ª–µ–¥—É—é—â–∏–µ 4 ‚Äî float)
    for i in range(1, len(b) - 4, 8):
        try:
            val = struct.unpack('<f', b[i:i+4])[0]  # little endian float32
            if np.isfinite(val):
                floats.append(val)
        except:
            pass
    return np.array(floats, dtype=np.float32)

# –¢–µ—Å—Ç
b = sample_df['embedding'].iloc[0]
emb = decode_fixed(b)
print("–≠–º–±–µ–¥–¥–∏–Ω–≥ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω, —Ñ–æ—Ä–º–∞:", emb.shape)
print("–ü–µ—Ä–≤—ã–µ 10 –∑–Ω–∞—á–µ–Ω–∏–π:", emb[:10])
print("–ú–∏–Ω/–º–∞–∫—Å:", emb.min(), emb.max())
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# –ü—É—Ç—å –∫ –º–æ–¥–µ–ª–∏
model_name = "cointegrated/rubert-tiny2"

# –ó–∞–≥—Ä—É–∂–∞–µ–º —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä –∏ –º–æ–¥–µ–ª—å
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# –ü–µ—Ä–µ–≤–æ–¥–∏–º –º–æ–¥–µ–ª—å –≤ eval —Ä–µ–∂–∏–º (–Ω–µ –æ–±—É—á–∞–µ–º)
model.eval()
```

```python
def get_embedding(text):
    """
    –ü–æ–ª—É—á–∞–µ—Ç —ç–º–±–µ–¥–¥–∏–Ω–≥ —Ç–µ–∫—Å—Ç–∞ —á–µ—Ä–µ–∑ rubert-tiny2
    """
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        # –ë–µ—Ä–µ–º CLS-—Ç–æ–∫–µ–Ω –∫–∞–∫ —ç–º–±–µ–¥–¥–∏–Ω–≥
        emb = outputs.last_hidden_state[:,0,:].squeeze(0)
    return emb.numpy()
```

```python
# –î–µ–∫–æ–¥–∏—Ä—É–µ–º –≤–µ—Å—å —Å—Ç–æ–ª–±–µ—Ü
sample_df['model_text_decoded'] = sample_df['model_text'].apply(lambda x: x.decode('utf-8'))

# –ü—Ä–æ–≤–µ—Ä–∏–º
sample_df['model_text_decoded'].head()

sample_texts = sample_df['model_text_decoded'].iloc[:3].tolist()

for i, t in enumerate(sample_texts):
    emb = get_embedding(t)
    print(f"–¢–µ–∫—Å—Ç {i}:", t[:50], "...")
    print(f"–≠–º–±–µ–¥–¥–∏–Ω–≥ {i} —Ñ–æ—Ä–º–∞:", emb.shape)
    print("–ü–µ—Ä–≤—ã–µ 5 —ç–ª–µ–º–µ–Ω—Ç–æ–≤:", emb[:5], "\n")
```

```python
import numpy as np
import torch
from tqdm import tqdm

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ–∫—Å—Ç–æ–≤ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ (–º–æ–∂–Ω–æ —É–º–µ–Ω—å—à–∏—Ç—å/—É–≤–µ–ª–∏—á–∏—Ç—å)
num_samples = 1000

# –ë–µ—Ä–µ–º —Å–ª—É—á–∞–π–Ω—É—é –ø–æ–¥–≤—ã–±–æ—Ä–∫—É
texts = sample_df['model_text_decoded'].sample(num_samples, random_state=42).tolist()

# –§—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        emb = outputs.last_hidden_state[:,0,:].squeeze(0)  # CLS —Ç–æ–∫–µ–Ω
    return emb.numpy()

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ —Å –ø—Ä–æ–≥—Ä–µ—Å—Å–±–∞—Ä–æ–º
embeddings = []
for t in tqdm(texts, desc="–ì–µ–Ω–µ—Ä–∞—Ü–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤"):
    emb = get_embedding(t)
    embeddings.append(emb)

embeddings = np.array(embeddings, dtype=np.float32)
print("–§–æ—Ä–º–∞ –∏—Ç–æ–≥–æ–≤–æ–π –º–∞—Ç—Ä–∏—Ü—ã —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤:", embeddings.shape)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏
np.save("./data/embeddings_sample.npy", embeddings)
print("–≠–º–±–µ–¥–¥–∏–Ω–≥–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ ./data/embeddings_sample.npy")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# –ü—Ä–∏–º–µ—Ä: —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 5 —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤
sim_matrix = cosine_similarity(embeddings[:5])
print(sim_matrix)
```

```python
from sklearn.cluster import KMeans

n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)
print("–ú–µ—Ç–∫–∏ KMeans:", labels[:10])
```

```python
import umap
import plotly.express as px
import pandas as pd

reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
emb_2d = reducer.fit_transform(embeddings)

viz_df = pd.DataFrame({
    "x": emb_2d[:,0],
    "y": emb_2d[:,1],
    "kmeans": labels,
    "text": texts
})

fig = px.scatter(viz_df, x='x', y='y', color='kmeans', hover_data=['text'])
fig.show()
```
